## Check Service Status

You can verify that the LDaCA services are running properly:

```python
import requests
import json

def check_ldaca_status():
    """Check if LDaCA services are running"""
    try:
        # Check LDaCA backend health
        health_url = 'http://localhost:8080/api/health'
        response = requests.get(health_url, timeout=5)
        
        if response.status_code == 200:
            print("‚úÖ LDaCA Backend: Running")
            try:
                health_data = response.json()
                print(f"   Status: {health_data.get('status', 'Unknown')}")
            except:
                print("   Status: Healthy")
        else:
            print(f"‚ö†Ô∏è  LDaCA Backend: HTTP {response.status_code}")
            
    except requests.exceptions.RequestException as e:
        print(f"‚ùå LDaCA Backend: Not accessible")
        print(f"   This is expected if running on ARM64/M1 Mac locally")
        print(f"   BinderHub deployment will work correctly on AMD64 architecture")
    
    try:
        # Check frontend access
        frontend_url = 'http://localhost:8080/'
        response = requests.get(frontend_url, timeout=5)
        
        if response.status_code == 200:
            print("‚úÖ LDaCA Frontend: Running")
        else:
            print(f"‚ö†Ô∏è  LDaCA Frontend: HTTP {response.status_code}")
            
    except requests.exceptions.RequestException as e:
        print(f"‚ùå LDaCA Frontend: Not accessible")
        print(f"   This is expected if running on ARM64/M1 Mac locally")
    
    print("\nüåê Access URL: /proxy/8080/")
    print("\nüìù Note: If services aren't running locally due to architecture")
    print("   mismatch, they will work correctly on BinderHub (AMD64)")

# Run status check
check_ldaca_status()
```
