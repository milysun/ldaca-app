```python
# LDaCA Service Startup (Run this first!)
import subprocess
import time
import os
import requests
from pathlib import Path

print("üöÄ Starting LDaCA Services...")

# Check if services are already running
def check_service(port, name):
    try:
        response = requests.get(f"http://localhost:{port}", timeout=2)
        return response.status_code < 500
    except:
        return False

# Check current service status
ldaca_running = check_service(443, "LDaCA Web App")
api_running = check_service(8001, "LDaCA API")

if ldaca_running and api_running:
    print("‚úÖ All LDaCA services are already running!")
else:
    print("üîÑ Starting LDaCA services...")
    
    # Start the service launcher in background
    if os.path.exists("/usr/local/bin/start-ldaca-services.sh"):
        try:
            # Run the startup script
            subprocess.Popen(["/usr/local/bin/start-ldaca-services.sh"], 
                           stdout=subprocess.PIPE, 
                           stderr=subprocess.PIPE)
            print("üîÑ Service launcher started...")
            
            # Wait a moment for services to start
            time.sleep(5)
            
            # Re-check services
            ldaca_running = check_service(443, "LDaCA Web App")
            api_running = check_service(8001, "LDaCA API")
            
            if ldaca_running:
                print("‚úÖ LDaCA Web App is running!")
            else:
                print("‚ö†Ô∏è  LDaCA Web App may still be starting...")
                
            if api_running:
                print("‚úÖ LDaCA API is running!")
            else:
                print("‚ö†Ô∏è  LDaCA API may still be starting...")
                
        except Exception as e:
            print(f"‚ö†Ô∏è  Error starting services: {e}")
            print("üîß You can manually start services later if needed")
    else:
        print("‚ö†Ô∏è  Service launcher script not found")

print(f"\nüåê Quick Links:")
print(f"  ‚Ä¢ LDaCA Web App: ../proxy/443/")
print(f"  ‚Ä¢ LDaCA API Docs: ../proxy/8001/docs")
print(f"  ‚Ä¢ API Health: ../proxy/8001/api/health")

# Check for startup marker from postBuild
if Path("/home/jovyan/.ldaca-start-marker").exists():
    print("\n‚úÖ PostBuild completed successfully")
```

## ‚è∞ Service Startup Information

**Important**: LDaCA services (nginx, FastAPI backend) are starting in the background and may take 10-15 seconds to become fully available after Jupyter loads.

### Service Status Indicators:
- ‚úÖ **Jupyter**: Ready immediately  
- üîÑ **LDaCA Services**: Starting in background (check `/tmp/supervisor.log` for details)
- üåê **Web App Access**: Available at [/proxy/443/](../proxy/443/) once services are ready
- üìö **API Access**: Available at [/proxy/8001/docs](../proxy/8001/docs) once services are ready

If LDaCA web app links don't work immediately, please wait a moment and try again!

```python
# Check LDaCA Service Status
import requests
import time
import os

def check_service_status():
    """Check if LDaCA services are ready"""
    services = {
        "LDaCA Web App (nginx)": "http://localhost:443",
        "LDaCA API (FastAPI)": "http://localhost:8001/api/health"
    }
    
    print("üîç Checking LDaCA service status...\n")
    
    for service_name, url in services.items():
        try:
            response = requests.get(url, timeout=3)
            if response.status_code == 200:
                print(f"‚úÖ {service_name}: Ready")
            else:
                print(f"üîÑ {service_name}: Starting (HTTP {response.status_code})")
        except requests.exceptions.ConnectionError:
            print(f"üîÑ {service_name}: Starting (not yet available)")
        except requests.exceptions.Timeout:
            print(f"‚è≥ {service_name}: Starting (timeout)")
        except Exception as e:
            print(f"‚ùì {service_name}: {str(e)}")
    
    # Check supervisor log if available
    log_file = "/tmp/supervisor.log"
    if os.path.exists(log_file):
        print(f"\nüìã Recent supervisor log entries:")
        with open(log_file, 'r') as f:
            lines = f.readlines()
            for line in lines[-5:]:  # Show last 5 lines
                print(f"   {line.strip()}")
    else:
        print("\n‚è≥ Supervisor log not yet available - services still starting")
    
    print(f"\nüåê Access links:")
    print(f"   ‚Ä¢ LDaCA Web App: /proxy/443/")
    print(f"   ‚Ä¢ LDaCA API: /proxy/8001/docs")

# Run the check
check_service_status()
```
